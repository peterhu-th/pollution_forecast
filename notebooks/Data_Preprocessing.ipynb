# 数据预处理 (Data Preprocessing)
本 Notebook 负责读取原始数据，进行缺失值处理、标准化、并在时间序列上切分为训练集(70%)、验证集(15%)和测试集(15%)。

### Cell 1: 库导入和路径设置

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

RAW_DIR = '../data/'
PROCESSED_DIR = '../data/processed/'
os.makedirs(PROCESSED_DIR, exist_ok=True)
```

### Cell 2: 读取与格式化数据

```python
file_path = os.path.join(RAW_DIR, 'data_merged.csv') 

try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"File not found: {file_path}")

df['datetime'] = pd.to_datetime(df['datetime'])
df = df.set_index('datetime')
df = df.sort_index()

numeric_cols = df.columns.tolist()
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print("Data info:")
print(df.info())
```

### Cell 3: 数据集切分与保存

```python
n = len(df)
train_size = int(n * 0.7)
val_size = int(n * 0.15)

train_df = df.iloc[0 : train_size]
val_df = df.iloc[train_size : train_size + val_size]
test_df = df.iloc[train_size + val_size :]

print(f"\nData split completed:")
print(f"Train: {len(train_df)} Samples: ({train_df.index.min()} to {train_df.index.max()})")
print(f"Val:   {len(val_df)} Samples: ({val_df.index.min()} to {val_df.index.max()})")
print(f"Test:  {len(test_df)} Samples: ({test_df.index.min()} to {test_df.index.max()})")

train_df.to_csv(os.path.join(PROCESSED_DIR, 'train_set.csv'))
val_df.to_csv(os.path.join(PROCESSED_DIR, 'val_set.csv'))
test_df.to_csv(os.path.join(PROCESSED_DIR, 'test_set.csv'))

print("Data Preprocessing Pipeline Ready.")
```
